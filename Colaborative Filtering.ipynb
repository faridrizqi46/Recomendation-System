```python
import pandas as pd
```

```python
rating = pd.read_csv('https://raw.githubusercontent.com/codeheroku/Introduction-to-Machine-Learning/master/Collaborative%20Filtering/dataset/ratings.csv')
rating.head()
```

```python
movie = pd.read_csv('https://raw.githubusercontent.com/codeheroku/Introduction-to-Machine-Learning/master/Collaborative%20Filtering/dataset/movies.csv')
movie.head()
```

```python
df = pd.merge(movie,rating)
```

```python
df.head()
```

```python
df.drop(columns=['genres','timestamp'], inplace=True)
```

```python
df.head()
```

```python
user_rating = df.pivot_table(index=('userId'), columns=['title'], values='rating')
user_rating
```

```python
# Remove Movie Where only have <= 10 Users who rated it , and fill rest NaN with 0
user_rating2 = user_rating.dropna(thresh=10, axis=1).fillna(0)
user_rating2
```

```python
corrMatrix = user_rating2.corr(method='pearson')
```

```python
def get_similar(movie_name,rating):
    similar_ratings = corrMatrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
romantic_lover = [("(500) Days of Summer (2009)",1),("Alice in Wonderland (2010)",1),("Aliens (1986)",5),("2001: A Space Odyssey (1968)",2)]
similar_movies = pd.DataFrame()
for movie,rating in romantic_lover:
    similar_movies = similar_movies.append(get_similar(movie,rating),ignore_index = True)

similar_movies.head(10)
```
